```python
def rank1(entry1, entry2):
    # entry is of the form [E,R,A,C,L]
    # Higher Effective -> Higher Relevance -> Less Aggressive -> Less Complex -> Neutral Length
    v1 = entry1[0]
    v2 = entry2[0]
    # return -1 entry1 item should be sorted before the entry2 item
    # check effectiveness
    if v1[0] > v2[0]:
        return -1
    elif v1[0] < v2[0]:
        return 1
    # check relevance
    if v1[1] > v2[1]:
        return -1
    elif v1[1] < v2[1]:
        return 1
    # check aggressiveness
    if v1[2] < v2[2]:
        return -1
    elif v1[2] > v2[2]:
        return 1
    # check complexity
    if v1[3] < v2[3]:
        return -1
    elif v1[3] > v2[3]:
        return 1  
    # check length - if both are not neutral, longer is preferred
    if v1[3] == 2 and v2[3]!=2:
        return -1
    elif v2[3] == 2 and v1[3]!=2:
        return 1
    elif v1[3] == 3 and v2[3]==1:
        return -1
    elif v2[3] == 3 and v1[3]==1:
        return 1
    return 0
```

```python
import functools
for hs in hs_cs_dict.keys():
    hs_cs_dict[hs] = sorted(hs_cs_dict[hs],key=functools.cmp_to_key(rank1))
    
```
